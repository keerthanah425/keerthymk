<h1 align=center>Modules and Packages</h1>

A Python module is like a single file with Python code inside. It has functions, statements, variables, and classes. You can bring it into other programs and use it there.

A package is a bunch of modules stored in a folder. Packages help us bundle several related modules into one place, using a shared name space. This makes managing and arranging our code simpler.

__init__.py file :This file can be empty or can contain an initialization code for your package. It signals to Python that the directory should be treated as a package.

```python
# Section 1: Importing and Using Functions from module1

import module1

module1.module1_func1()
module1.module1_func2()

print(module1.__name__)
```

```python
# Section 2: Importing with Alias and Function Call

import module1 as m1

m1.module1_func1()
```

```python
# Section 3: Importing Specific Functions and Calling Them

from module1 import module1_func1, module1_func2

module1_func1()
module1_func2()
```

```python
# Section 4: Importing with Alias and Renaming a Function

from module1 import module1_func1 as myfun

myfun()
```

```python
# Section 5: Importing Function from package1.module2

import package1.module2 as a

a.module2_func2()
```

```python
# Section 6: Importing and Calling Function from subModule1

import package1.subPackage1.subModule1

pacakage1.subPackage1.subModule1.subfunc1()
```

```python
# Section 7: Importing with Alias and Calling Function from subModule1

import package1.subPackage1.subModule1 as al

al.subfunc1()
```

```python
# Section 8: Importing and Calling Function from subModule1 using from...import...

from package1.subPackage1 import subModule1

subModule1.subfunc2()
```

```python
# Section 9: Importing with Alias and Calling Function from subModule1 using from...import...

from package1.subPackage1 import subModule1 as sb

sb.subfunc2()
```

```python
# Section 10: Importing Function from subModule2

from package1.subPackage1.subModule2 import subfunc2

subfunc2()
```

```python
# Section 11: Importing Multiple Functions from module1 and Calling Them

from package1.module1 import module1_func2, module1_func1

module1_func2()
module1_func1()
```

```python
# Section 12: Importing All Functions from module1 and Calling Them

from package1.module1 import *

module1_func2()
module1_func1()
```

```python
# Section 13: Using __all__ to Control Importing

from pacakage1 import *

module1_func1() # this wont be imported because of __all__
```

```python
# Example using the math module
import math
print(math.pi)
print(math.sin(math.pi / 2))
```

```python
# Example using the json module
import json
data = {
    "name": "John Doe",
    "age": 30,
    "city": "New York"
}
json_data = json.dumps(data)
print(json_data)
```

```python
# Example using the datetime module
import datetime
now = datetime.datetime.now()
print(now)
print(now.year)
print(now.month)
print(now.day)
```

```python
# Example using the re module
import re
text = "The quick brown fox jumps over the lazy dog."
result = re.search("fox", text)
print(result.start(), result.end(), result.group())
```

```python
# Example using the random module
import random
print(random.randint(1, 100))
print(random.choice([1, 2, 3, 4, 5]))
```

```python
import zlib
data = b'Hello everyone'
compressed = zlib.compress(data)
print ("Compressed:",compressed)
decompressed = zlib.decompress(compressed)
print ("Decompressed:", decompressed)
```

```python
from string import Template

template = Template("Hello, my name is $name and I am $age years old.")
message = template.substitute(name="Charlie", age=35)
print(message)
```
