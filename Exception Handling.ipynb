### Handling Specific Exceptions

```python
try:
    # Code that may raise an exception
    x = int(input("Enter a number: "))
    result = 10 / x
except ValueError:
    # Handle ValueError
    print("Error: Invalid input! Please enter a valid number.")
except ZeroDivisionError:
    # Handle ZeroDivisionError
    print("Error: Division by zero!")
```

```python
try:

    even_numbers = [2,4,6,8]
    print(even_numbers[5])

except ZeroDivisionError:
    print("Denominator cannot be 0.")

except IndexError:
    print("Index Out of Bound.")
```

### The try-except-else Block

The else block is executed if no exception occurs in the try block.

```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ValueError:
    # Handle ValueError
    print("Error: Invalid input! Please enter a valid number.")
except ZeroDivisionError:
    # Handle ZeroDivisionError
    print("Error: Division by zero!")
else:
    print("Result:", result)
```

```python

def funct(a , b):
    try:
        c = ((a+b) / (a-b))
    except ZeroDivisionError:
        print ("a/b result in 0")
    else:
        print (c)
funct(2.0, 3.0)
funct(3.0, 3.0)
```

The as keyword is used in the except block to assign the caught exception object to a variable

```python
try:
    num = int(input("Enter a number: "))
    reciprocal = 1/num
except (ZeroDivisionError,ValueError) as a:
    print(a)
else:

    print(reciprocal)
```

```python
def calculate_average(numbers):
    try:
        if not numbers:
            raise ValueError("List is empty!")
        avg = sum(numbers) / len(numbers)
    except ValueError as ve:
        print("Error:", ve)
    except TypeError as te:
        print("Error:", te)
    else:
        print("Average:", avg)

# Test cases
calculate_average([1, 2, 3, 4, 5])  # Output: Average: 3.0
calculate_average([])              # Output: Error: List is empty!
calculate_average([1, 2, 'three']) # Output: Error: unsupported operand type(s) for +: 'int' and 'str'
```

### The try-except-finally Block

The finally block is always executed

```python
try:
    numerator = 10
    denominator = 0

    result = numerator/denominator

    print(result)
except:
    print("Error: Denominator cannot be 0.")

finally:
    print("This is finally block.")
```

```python
try:
    file = open("example.txt", "r")
    # Perform operations on the file
except FileNotFoundError:
    print("Error: File not found!")
finally:
     print("Finally block")
      # Always close the file, even if an exception occurs
```

### Raising Exceptions

You can manually raise exceptions using the raise keyword

```python
x = -1
if x < 0:
    raise ValueError("Invalid value: x must be non-negative")
```

```python
try:
    raise NameError("Hi there")
except NameError as name:
    print ("An exception")
    print(name)
```

```python
num = [3, 4, 5, 7]
if len(num) > 3:
    raise Exception( f"Length of the given list must be less than or equal to 3 but is {len(num)}" )
```

```python
def factorial(n):
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a positive integer")
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Test cases
try:
    print(factorial(5))   # Output: 120
    print(factorial(-5))  # Raises ValueError: Input must be a positive integer
except ValueError as ve:
    print("Error:", ve)
```

### Custom Exceptions

```python
class CustomError(Exception):
    pass

# Raise CustomError
raise CustomError("This is a custom error message")
```

```python
try:
    # Some code that may raise CustomError
    raise CustomError("This is a custom error message")
except CustomError as e:
    print("Caught CustomError:", e)
```

```python
# define Python user-defined exceptions
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass


number = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")

except InvalidAgeException:
    print("Exception occurred: Invalid Age")
```

### The assert statement

the assert statement is used to ensure that the denominator y is not zero before performing division. If the condition specified in the assert statement evaluates to False, an AssertionError is raised with the specified error message.

```python
# Example of using assert statement for debugging
def divide(x, y):
    assert y != 0, "Division by zero!"  # Check if y is not zero
    return x / y

# Test cases
print(divide(10, 2))  # Output: 5.0
print(divide(10, 0))  # Raises AssertionError: Division by zero!
```

```python
def square_root( Number ):
    assert ( Number < 0), "Give a positive integer"
    return Number**(1/2)

#Calling function and passing the values
print( square_root( 36 ) )
print( square_root( -36 ) )
```
