@classmethod is used when a method needs to work with the class itself rather than a specific object.

```python
class Student:

    # Normal method
    # 'self' refers to the object that calls this method
    def say_hello(self):
        print("Hello from the student object!")


# Create an object of the class
s1 = Student()

# Call the method using the object
s1.say_hello()
```

```python
class Student:

    # Class method
    # 'cls' refers to the class itself
    @classmethod
    def say_hello(cls):
        print("Hello from the Student class!")


# Call the method using the class name
Student.say_hello()

# self = object
# cls  = class
```

### Encapsulation

Encapsulation is the bundling of data and methods that operate on that data, restricting direct access to some of the object's components and hiding the internal state.

```python
class Bank:
    __balance = 25000
    __pin = 1234

    def setBalance(self, pin, balance):  # Setter function
        if self.__pin != pin:
            print("Wrong pin Entered")
        else:
            self.__balance = balance

    def getBalance(self, pin):   # getter function
        if self.__pin != pin:
            print("Wrong pin Entered")
            return
        else:
            return self.__balance
```

```python
personA=Bank()  # Real bank account owner
personA.setBalance(1234,15000)
result=personA.getBalance(1234)
print(f"The returned balanceis {result}")
```

### Inheritance


Inheritance is a mechanism in object-oriented programming where a new class (subclass) is created by deriving properties and behaviors from an existing class (superclass).

#### Single level Inheritance

```python
class BaseClass:
    def funct1(self):
        print("This is BaseClass function")

class DerivedClass(BaseClass):
    def funct2(self):
        print("This is DerivedClass function")

obj=DerivedClass()
obj.funct1()
obj.funct2()
```

#### Multi level Inheritance

```python
class GrandParent:
    def funct1(self):
        print("This is GrandParent function")

class Parent(GrandParent):
    def funct2(self):
        print("This is Parent function")

class Child(Parent):
    def funct3(self):
        print("This is Child function")

obj=Child()
obj.funct1()
obj.funct2()
obj.funct3()
```

#### Multiple inheritance

```python
class Father:
    def funct1(self):
        print("This is Father class function")

class Mother:
    def funct2(self):
        print("This is Mother class function")

class Child(Father,Mother):
    def funct3(self):
        print("This is Child class function")

obj=Child()
obj.funct1()
obj.funct2()
obj.funct3()
```

#### Hierarchical Inheritance

```python
class Parent:
    def funct1(self):
        print("This is Parent class function")

class Child1(Parent):
    def funct2(self):
        print("This is Child1 class function")

class Child2(Parent):
    def funct3(self):
        print("This is Child2 class function")

c2=Child2()
c2.funct1()
c2.funct3()
```

#### super() function

to access the superclass method from the subclass we use the super() function

```python
class GrandParent:
    def funct1(self):
        print("This is GrandParent function")

class Parent(GrandParent):
    def funct2(self):
        print("This is Parent function")
        super().funct1()

class Child(Parent):
    def funct3(self):
        print("This is Child function")
        super().funct2()


obj=Child()
obj.funct3()
```

#### Polymorphism

a core principle of object-oriented programming (OOP), empowers methods to exhibit different behaviors based on the object they are acting upon.

```python
# Demonstrating polymorphism with the built-in function len()
print(len("Hello, World!"))  # Length of a string
print(len([1, 2, 3, 4, 5]))  # Length of a list
```

```python
# polymorphism with addition operator
result = 5 + 3
print(result)

result = "Hello, " + "World!"
print(result)

list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = list1 + list2
print(result)
```

```python
# Class polymorphism
class Dog:
    def __init__(self, breed, name):
        self.breed = breed
        self.name = name

    def sound(self):
        print("Woof!")

class Cat:
    def __init__(self, breed, name):
        self.breed = breed
        self.name = name

    def sound(self):
        print("Meow!")

class Bird:
    def __init__(self, species, name):
        self.species = species
        self.name = name

    def sound(self):
        print("Tweet!")

dog1 = Dog("Labrador", "Buddy")  # Create a Dog class
cat1 = Cat("Siamese", "Whiskers")  # Create a Cat class
bird1 = Bird("Parrot", "Polly")  # Create a Bird class

for animal in (dog1, cat1, bird1):
    animal.sound()
```

```python
# Polymorphism via inheritance (method overriding)
class Animal:
    def __init__(self, species, name):
        self.species = species
        self.name = name

    def sound(self):
        print("Make a sound!")

class Dog(Animal):
    pass

class Cat(Animal):
    def sound(self):
        print("Meow!")

class Bird(Animal):
    def sound(self):
        print("Tweet!")

dog1 = Dog("Dog", "Buddy")
cat1 = Cat("Cat", "Whiskers")
bird1 = Bird("Bird", "Polly")

for x in (dog1, cat1, bird1):
    print(x.species)
    print(x.name)
    x.sound()
```

#### Abstraction

the concept of hiding the complex implementation details of a piece of code while exposing only the necessary functionalities to the user

```python
# # Implementing Abstraction in Python

# Abstraction is a key concept in object-oriented programming (OOP) that involves hiding the complex implementation details of a system and exposing only the necessary parts. It allows us to create more user-friendly and maintainable code. In Python, abstraction can be achieved using abstract classes and interfaces.

# ## Abstract Classes

# An abstract class is a class that cannot be instantiated and often includes one or more abstract methods. Abstract methods are methods that are declared but contain no implementation. Python's `abc` module provides tools for defining abstract base classes.

from abc import ABC, abstractmethod

# Define an abstract class using the ABC module
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
```

```python
# ## Concrete Classes

# Concrete classes are subclasses of the abstract class that provide implementations for the abstract methods. Let's create a couple of concrete classes: `Rectangle` and `Circle`.

import math

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

# ## Using the Classes

# We can now create instances of `Rectangle` and `Circle` and call their methods to see abstraction in action.

# Create a rectangle and a circle
rect = Rectangle(3, 4)
circ = Circle(5)

# Print the area and perimeter of the rectangle
print(f"Rectangle: Area = {rect.area()}, Perimeter = {rect.perimeter()}")

# Print the area and perimeter of the circle
print(f"Circle: Area = {circ.area()}, Perimeter = {circ.perimeter()}")

# ## Explanation

# In this example, the `Shape` class is an abstract class with two abstract methods: `area` and `perimeter`. The `Rectangle` and `Circle` classes inherit from `Shape` and provide concrete implementations of these methods. This setup allows us to work with shapes in a general way, without needing to know the specifics of how each shape calculates its area and perimeter.

# By using abstraction, we can create a system that's easier to extend and maintain. If we need to add new shapes in the future, we simply create new subclasses of `Shape` and implement the required methods, without modifying the existing codebase.

# ## Conclusion

# Abstraction helps in managing complexity by hiding unnecessary details and exposing only the essential features of an object. In Python, abstraction is typically achieved using abstract base classes and interfaces, which define a common interface for a group of related classes, while allowing the concrete classes to implement the details.
```
